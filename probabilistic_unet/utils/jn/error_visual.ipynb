```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
df = pd.read_csv("./csvs/new_geco.csv")
```

```python
df
```

```python
plt.rcParams.update({'font.size': 50})
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
linewidth = 10

# plt.rc('font', family='serif', serif='cm10')
# plt.rc('text', usetex=True)
# plt.rcParams['text.latex.preamble'] = [r'\boldmath']
```

```python
fig, ax1 = plt.subplots(figsize=(20,16))

# Plotting the coefficient changes on the primary y-axis
color = 'tab:blue'
ax1.set_xlabel('Epochs')
ax1.set_ylabel('log$p(y_s|x, z)$', color=color)
ax1.plot(df["volvo_onTheFly_new_GECO_lagrange_exp16_zero_ema - Reconstruction training loss"], color=color, linewidth=linewidth)
line1 = ax1.axhline(y = 1.9, color = 'r', linestyle = '-', label = "$goal_{seg}$", linewidth=linewidth) 
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(axis = "y")



ax2 = ax1.twinx()
color = 'tab:green'
ax2.set_ylabel('$\lambda_S$', color=color)
ax2.plot(df["volvo_onTheFly_new_GECO_lagrange_exp16_zero_ema - lambda seg coeficient"], color=color, linewidth=linewidth)
ax2.tick_params(axis='y', labelcolor=color)


ax1.legend()
plt.savefig('./pics/zero_ema_myGECO_reco.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(20,16))

# Plotting the coefficient changes on the primary y-axis
color = 'tab:blue'
ax1.set_xlabel('Epochs')
ax1.set_ylabel('log$p(y_f|x, z)$', color=color)
ax1.plot(df["volvo_onTheFly_new_GECO_lagrange_exp16_zero_ema - Regression training loss"], color=color, linewidth=linewidth)
ax1.axhline(y = 0.05, color = 'r', linestyle = '-', label = "$goal_{fri}$", linewidth=linewidth) 
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(axis = "y")



ax2 = ax1.twinx()
color = 'tab:green'
ax2.set_ylabel('$\lambda_f$', color=color)
ax2.plot(df["volvo_onTheFly_new_GECO_lagrange_exp16_zero_ema - lambda fri coeficient"], color=color, linewidth=linewidth)
ax2.tick_params(axis='y', labelcolor=color)

ax1.legend()
plt.savefig('./pics/zero_ema_myGECO_regression.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
fig, ax1 = plt.subplots()

# Plotting the coefficient changes on the primary y-axis
color = 'tab:blue'
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Reconstruction loss', color=color)
ax1.plot(df["volvo_onTheFly_new_GECO_lagrange_exp16_zero_ema - KL training loss"], color=color)
ax1.tick_params(axis='y', labelcolor=color)


ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Lambda s', color=color)
ax2.plot(df["volvo_onTheFly_new_GECO_lagrange_exp16_zero_ema - lambda kl coeficient"], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Title and showing the plot
plt.title('Coefficient Changes and Loss Function')
plt.grid()
plt.show()
```
