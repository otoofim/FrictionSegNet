```python
def singleFrame(vid_add, indexFrame):
    
    filename = glob.glob(os.path.join(vid_add,"*.mp4"))[0]
    cap = cv2.VideoCapture(filename)
    cap.set(1, indexFrame)
    ret, frame = cap.read()


    cap.release()
    cv2.destroyAllWindows()
    
    return cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
```
